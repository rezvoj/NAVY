```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

#### Function to generate IFS model for the given affine transformations

```python
def generate_ifs_model(matrices: list[np.ndarray], add_vectors: list[np.ndarray], num_iter: int):
    history = [np.array([0.0, 0.0, 0.0])]
    for _ in range(num_iter):
        # Pick one affine transformation randomly
        rand_idx = np.random.randint(len(matrices))
        matrix, add_vector = matrices[rand_idx], add_vectors[rand_idx]
        # Apply the transformation to the last vector
        history.append(matrix.dot(history[-1]) + add_vector)
    return history
```

#### Function to visualize the IFS model

```python
def visualize_ifs_model(vector_history: list[np.ndarray]):
    points_to_plot = np.array(vector_history)
    x, y, z = points_to_plot[:, 0], points_to_plot[:, 1], points_to_plot[:, 2]
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z, s=2, alpha=0.8, marker='.')
    plt.show()
```

#### Visualize IFS models for the specified affine transformations

```python
matrices = [
    np.array([[0.00,  0.00,  0.01], [0.00,  0.26,  0.00], [0.00,  0.00,  0.05]]),
    np.array([[0.20, -0.26, -0.01], [0.23,  0.22, -0.07], [0.07,  0.00,  0.24]]),
    np.array([[-0.25,  0.28,  0.01], [ 0.26,  0.24, -0.07], [ 0.07,  0.00,  0.24]]),
    np.array([[0.85,  0.04, -0.01], [-0.04,  0.85,  0.09], [ 0.00,  0.08,  0.84]])
]
add_vectors = [
    np.array([0.00, 0.00, 0.00]),
    np.array([0.00, 0.80, 0.00]),
    np.array([0.00, 0.22, 0.00]),
    np.array([0.00, 0.80, 0.00])
]
ifs_model = generate_ifs_model(matrices, add_vectors, 1000000)
visualize_ifs_model(ifs_model)
```
